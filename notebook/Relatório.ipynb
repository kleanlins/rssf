Cada nó possuem valores em X e em Y que são usados para calcular o alcance de cada nó a outro, usando a _distância euclidiana h² = a² + b²_, fazendo assim o cálculo da hipotenusa do triângulo que forma entre eles, que é a distância entre os nós propriamente dita, de forma precisa e visual tendo o alcance dos nós na rede. 


```python
import matplotlib.pyplot as plt
import sys
import numpy as np
import math as mt
sys.path.insert(0, '../model')

# user made models
# import host
# import package
# import router
import utils

## AREA DESCRIPTION ##

hosts_quantity = 10

# values below corresponds to a measurement in kilometers
width = 15
height = 15

# get arrays containing coordinates
x, y = utils.generate_coordinates(hosts_quantity, width, height)

# figure dimensions
plt.axis([0, width, 0, height])

# figure labels
plt.xlabel("in Km")
plt.ylabel("in Km")

# plotting triangles
plt.plot(x, y, 'k^')

n = []
# plotting hosts names
for i in range(hosts_quantity):
    plt.annotate(i+1, (x[i]+0.1, y[i]+0.1))
    n.append(i+1)

print("em x:",x ,"\nem y:" ,y) 

adj_list = list()

for j in range(hosts_quantity - 1):
    for k in range(hosts_quantity - 1):
        a = abs(x[j+1] - x[k])
        b = abs(y[j+1] - y[k])
        h = mt.sqrt(mt.pow(a,2) + mt.pow(b,2))
    
        if(h<=3.9 and h != 0 and x[j+1] != x[j] and y[j+1]!=y[j]):
            print(n[j+1]," Está no alcance de ", n[k])
            adj_list.append(n[j+1])
            adj_list.append(n[k])
        #else:
           # print(x[j],y[j]," Não está no alcance de ", x[k+1],y[k+1])

#plot the circles with a fixed area of 25000, what results in a Radius of 3.2 aprox.
r = mt.sqrt((22500/np.pi))
area = np.pi * (r*r)

# plotting estimated area coverage
plt.scatter(x, y, s=area, alpha=0.11)

plt.show()
```
